# Parte 1: Carga y Limpieza (Pandas y NumPy)

1. Importa las librerías necesarias (pandas, numpy, matplotlib.pyplot).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Lee el archivo CSV desde la URL proporcionada y guárdalo en un DataFrame.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)
```

3. Imprime las últimas 5 filas del dataset y genera un resumen estadístico descriptivo de las columnas numéricas.

```python
print("Últimas 5 filas del dataset")
display(df.tail())

print("Resumen estadístico de las columnas numéricas")
display(df.describe())
```

4. Eliminar filas donde las dimensiones físicas (x, y, z) sean exactamente 0

```python
df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]
```

5. Crea una nueva columna llamada volumen_aprox calculando el volumen espacial de cada pieza (multiplicando las columnas x * y * z).

```python
df['volumen_aprox'] = df['x'] * df['y'] * df['z']
```

Mostrar las primeras filas para verificar la nueva columna

```python
print("volumen_aprox")
df.head()
```

# Parte 2: Visualización de Datos (Matplotlib)

1. Gráfico de Caja (Boxplot): Muestra la distribución estadística de la columna price (precio) para identificar visualmente si existen valores atípicos (outliers).

```python

plt.figure()
plt.boxplot(df['price'])
plt.title('Distribución del precios de los diamantes')
plt.xlabel("Total de Diamantes")
plt.ylabel("Precio")
plt.show()
```

2. Gráfico de Dispersión (Scatter): Grafica el volumen_aprox en el eje X contra el price en el eje Y.

```python


plt.figure()

plt.scatter(df['volumen_aprox'], df['price'])
plt.title('Relación entre Volumen Aproximado y Precio')
plt.xlabel('Volumen Aproximado')
plt.ylabel('Precio')

plt.show()
```

3. Gráfico de Barras: Muestra la cantidad total de piezas que existen para cada tipo de corte (cut).

```python

conteo_cortes = df['cut'].value_counts()

plt.figure()
plt.bar(conteo_cortes.index, conteo_cortes.values)
plt.title('Cantidad de Piezas por Tipo de Corte')
plt.xlabel('Tipo de Corte')
plt.ylabel('Cantidad Total de Piezas')

plt.show()
```
